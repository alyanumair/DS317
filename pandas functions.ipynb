use to access rows and columns and extract a new filtered dataset, when given rows and columns it will only shows that specific row and that column<br>

```python

#It shows row number 10
data.loc[10]

#It shows row number 10 in a tabular form
data.loc[[10]]

#It shows row number 10 , 17 and 35
data.loc[[10 , 17 , 35]]

#It wil show row number 15 and OrderID column
data.loc[[15], ["OrderID"]]
```

# copy( )

This method will copy the current dataset. It takes only one optional argument <b>"deep"</b> which is by default true it creates a deep copy of the dataset, any changes in the orignal dataset will not effect in the copied dataset and vice verse.
If we set deep to false it will create a shallow copy and any change in the orignal dataset will reflect in the coped dataset and vice verse

```python


deep = data.copy()

shallow = data.copy(deep = False)



```

# Rename

This mode is us to rename rows and columns in a dataset. You have to pass an dictionary argument in which old index/labels are keys and the new index/label are values. There are also others ways to do this. By defalut after renaming it generate a new renamed dataframe but if we want to rename in the given dataframe we have to set inplace to True by defaul its false

```python
old = {
    
    "OrderID" : "PublicID",
    "Quantity" : "No."
}

deep.rename(columns = old , inplace = True)

deep

```
